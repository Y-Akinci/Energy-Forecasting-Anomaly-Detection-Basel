```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
import seaborn as sns
```

```python
root_directory = os.path.dirname(os.getcwd())
print(root_directory)

data_path_15min = os.path.join(root_directory, "Data", "Merged Data", "merged_strom_meteo_15min.csv")
data_path_30min = os.path.join(root_directory, "Data", "Merged Data", "merged_strom_meteo_30min.csv")


dataset_15min_intervall = pd.read_csv(data_path_15min, sep=";", decimal=",", encoding="latin1")
dataset_30min_intervall = pd.read_csv(data_path_30min, sep=";", decimal=",", encoding="latin1")
```

```python
processed_data_path = os.path.join(root_directory, "Data", "processed_merged_features.csv")

processed_data = pd.read_csv(processed_data_path, sep=";", decimal=",", encoding="latin1")
```

```python
# Zeilen löschen, welche keinen Wert bei der Zielvariable "Stromverbrauch" haben
dataset_15min_intervall.dropna(subset="Stromverbrauch", inplace=True)
```

### Datentypen

```python
# Datum und Zeitwert in datetime umwandeln
dataset_15min_intervall["DateTime"] = pd.to_datetime(dataset_15min_intervall["DateTime"], utc=True)

# "Date" und "Time" als eigene Spalten erstellen
dataset_15min_intervall["Date"] = dataset_15min_intervall["DateTime"].dt.strftime("%Y-%m-%d")
dataset_15min_intervall["Time"] = dataset_15min_intervall["DateTime"].dt.strftime("%H-%M-%S")

# DateTime in index umwandeln
dataset_15min_intervall.set_index("DateTime", inplace=True)
```

```python
# Datentyp ändern, Zeitwerte
dataset_15min_intervall["Date"] = pd.to_datetime(dataset_15min_intervall["Date"], format="%Y-%m-%d")
dataset_15min_intervall["Time"] = pd.to_datetime(dataset_15min_intervall["Time"], format="%H-%M-%S")
```

```python
# Datentyp ändern, integers
int_cols = [
    "Jahr", "Monat", "Tag", "Wochentag", "Tag des Jahres",
    "Quartal", "Woche des Jahres",
]

for col in int_cols:
    dataset_15min_intervall[col] = pd.to_numeric(
        dataset_15min_intervall[col], errors="coerce"
    ).astype("Int64") 
```

```python
# Datentypen ändern, float64
float_cols = [
    'Stromverbrauch',
    'Lufttemperatur 2 m Ã¼. Boden',
    'Lufttemperatur 2 m Ã¼. Gras',
    'Lufttemperatur BodenoberflÃ¤che',
    'Chilltemperatur',
    'relative Luftfeuchtigkeit',
    'Taupunkt 2 m Ã¼. Boden',
    'Dampfdruck 2 m Ã¼. Boden ',
    'Luftrdruck auf BarometerhÃ¶he',
    'Luftdruck reduziert auf Meeresniveau mit StandardatmosphÃ¤re (QNH)',
    'Luftdruck reduziert auf Meeresniveau (QFF)',
    'BÃ¶enspitze (SekundenbÃ¶e)',
    'Windgeschwindigkeit vektoriell; Zehnminutenmittel in m/s',
    'Windgeschwindigkeit skalar; Zehnminutenmittel in m/s',
    'Windrichtung; Zehnminutenmittel',
    'Windgeschwindigkeit; Zehnminutenmittel in km/h',
    'BÃ¶enspitze (3-SekundenbÃ¶e); Maximum in m/s',
    'BÃ¶enspitze (SekundenbÃ¶e); Maximum in km/h',
    'BÃ¶enspitze (3-SekundenbÃ¶e); Maximum in km/h',
    'Niederschlag; Zehnminutensumme',
    'Globalstrahlung; Zehnminutenmittel',
    'Diffusstrahlung; Zehnminutenmittel',
    'Langwellige Einstrahlung; Zehnminutenmittel',
    'Sonnenscheindauer; Zehnminutensumme', 'Grundversorgte Kunden', 'Freie Kunden'
]

for col in float_cols:
    dataset_15min_intervall[col] = pd.to_numeric(dataset_15min_intervall[col], errors="coerce").astype("float64")
```

```python
# Datentypen für KAtegoriale variablen
categorical_features = [
    "Jahr",
    "Monat",
    "Tag",
    "Wochentag",
    "Quartal",
    "Woche des Jahres"
]

for col in categorical_features:
    dataset_15min_intervall[col] = dataset_15min_intervall[col].astype("category")
```
